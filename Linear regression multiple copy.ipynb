```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
%matplotlib inline
```

```python
# reading the data from csv file 
dataset = pd.read_csv('Advertising.csv')
# showing the data 
dataset.head()
```

```python
# separting data 
x = dataset[['TV', 'Radio', 'Newspaper']].values
y = dataset['Sales'].values
```

```python
# verification des dimesions de x et y
print(x.shape)
print(y.shape)
#  redimensionnement de y
y = y.reshape(y.shape[0], 1)
print(y.shape)
```

```python
#  creation de la matrice 
X = np.hstack((x, np.ones((x.shape[0], 1))))
print(X.shape)
# affichage des 5 premiere valeur 
print(X[:5])

```

```python
#  initialization de theta avec 4 element 
theta = np.random.randn(4,1)
theta
```

```python
# creation de model avec multiplication des valeur avec le vecteur theta
def model(X , theta):
    return np.dot(X, theta)
```

```python
def cost_function(X, y, theta):
    m = len(y)
    return 1/(2*m) * np.sum((model(X, theta) - y)**2)
```

```python
cost_function(X, y, theta)
```

```python
def grad(X, y, theta):
    m = len(y)
    return 1/m * X.T.dot(model(X, theta) - y)
```

```python
def gradient_descent(X, y, theta, learning_rate, n_iterations):
    
    cost_history = np.zeros(n_iterations) # cration d'un tableau de stockage pour enregistrer l'volution du Cout du modele
    
    for i in range(0, n_iterations):
        theta = theta - learning_rate * grad(X, y, theta) # mise a jour du parametre theta (formule du gradient descent)
        cost_history[i] = cost_function(X, y, theta) # on enregistre la valeur du Cout au tour i dans cost_history[i]
        
    return theta, cost_history
```

```python
n_iterations = 1000000
learning_rate = 0.00001

theta_final, cost_history = gradient_descent(X, y, theta, learning_rate, n_iterations)
```

```python
# cration d'un vecteur prdictions qui contient les prdictions de notre modele final
predictions = model(X, theta_final)

theta_final
```

```python
#%matplotlib notebook
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(x[:,0], x[:,1], x[:,2], c=y)
ax.scatter(x[:,0], x[:,1], x[:,2], c=predictions)
# affiche les noms des axes
ax.set_xlabel('TV')
ax.set_ylabel('RADIO')
ax.set_zlabel('Newspaper')
ax.set_title("representation 4d des resultat")
plt.show()
```
